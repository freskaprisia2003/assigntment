## Data Preparation

```python
import pandas as pd
import matplotlib.pyplot as plt
df = pd.read_csv("data_monev.csv", skiprows=[1, 4, 8])
df.set_index("Angkatan", inplace=True)
df = df.apply(pd.to_numeric, errors='coerce')
```

## Data Visualization

#### Jumlah peserta berdasarkan jenjang pendidikan

```python
plt.figure(figsize=(8, 5))
df.loc["S1":"S2"].T.plot(kind="bar", stacked=True, colormap="viridis", figsize=(8, 5))
plt.title("Distribusi Jenjang Pendidikan per Angkatan")
plt.xlabel("Angkatan")
plt.ylabel("Jumlah Peserta")
plt.xticks(rotation=0)
plt.legend(title="Jenjang Pendidikan")
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.savefig("jenjang_pendidikan.png")
plt.show()
```

#### Insight ####
Mayoritas peserta Pengajar Muda berasal dari jenjang pendidikan S1, dengan proporsi yang sangat dominan di setiap angkatan. Meskipun ada peserta dengan jenjang S2, jumlahnya jauh lebih sedikit. Ini menunjukkan bahwa program lebih banyak menarik lulusan S1 dibandingkan lulusan S2, kemungkinan karena faktor usia, kesiapan lapangan, atau persyaratan program.

### Jumlah peserta berdasarkan jenis kelamin

```python
plt.figure(figsize=(8, 5))
df.loc["Laki-laki":"Perempuan"].T.plot(kind="bar", stacked=True, colormap="coolwarm", figsize=(8, 5))
plt.title("Distribusi Jenis Kelamin per Angkatan")
plt.xlabel("Angkatan")
plt.ylabel("Jumlah Peserta")
plt.xticks(rotation=0)
plt.legend(title="Jenis Kelamin")
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.savefig("jenis_kelamin.png")
plt.show()
```

#### Insight ####
Dari data ini terlihat bahwa jumlah peserta perempuan selalu lebih banyak dibandingkan laki-laki di setiap angkatan. Hal ini menunjukkan adanya ketertarikan atau keterlibatan yang lebih besar dari perempuan dalam program Pengajar Muda. Bisa jadi faktor sosial, minat dalam bidang pendidikan, atau ketersediaan kandidat perempuan lebih tinggi dibanding laki-laki.
